```python
twtr=k[["Twtr.","Name","Annotations"]]
```

I am unsing NLTK packages util.demo_liu_hu_lexicon function to find polarity of emoji Names. It is very simple function.
I uses scoring system to find polarity. It has a dictionary of positive, negative and neutral words. Function uses it to check if word in sentence are positive, negative or neutral. If word is not the the list, it will be tag it as neutral.

```python
twtr['Polarity'] = twtr['Name'].apply(lambda x: sentiment.util.demo_liu_hu_lexicon(x, plot=False))
```

This may not be very accurate, with this method. One way is to improve dictionary of positive, negative and neutral words by adding more words.
